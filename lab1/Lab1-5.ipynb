3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
сhildren = df[df["Age"] < 16]
adults = df[df["Age"] >= 16]
child_surv = сhildren[сhildren["Survived"] == 1]
adults_surv = adults[adults["Survived"] == 1]
print("Процент выживаемости детей: ", len(child_surv) / (len(сhildren) / 100), "%", sep='')
print("Процент выживаемости взрослых: ", len(adults_surv) / (len(adults) / 100), "%", sep='')
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
сhildren = df[df["Age"] < 16]
adults = df[df["Age"] >= 16]
man = adults[adults["Sex"] == "male"]
woman = adults[adults["Sex"] == "female"]
child_surv = сhildren[сhildren["Survived"] == 1]
man_surv = man[man["Survived"] == 1]
woman_surv = woman[woman["Survived"] == 1]
percent_child = len(child_surv) / (len(сhildren) / 100)
percent_man = len(man_surv) / (len(man) / 100)
percent_woman =  len(woman_surv) / (len(woman) / 100)
if percent_man < percent_child and percent_man < percent_woman:
  print("YES, man: ", percent_man, ", woman: ", percent_woman, ", children: ", percent_child, sep='')
else:
  print("NO, man: ", percent_man, ", woman: ", percent_woman, ", children: ", percent_child, sep='')
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
young_man = df[(df['Sex'] == 'male') & (df['Age'] >= 18) & (35 > df['Age'])]
old_man = df[df['Age'] >= 35]

percent_young = young_man['Survived'].mean() * 100
percent_old = old_man['Survived'].mean() * 100

if percent_young > percent_old:
  print('молодым выжить легче')
elif percent_young < percent_old:
  print('взрослым выжить легче')
else:
  print('без разницы')
print('процент молодых:', percent_young)
print('процент взрослых:', percent_old)
````
